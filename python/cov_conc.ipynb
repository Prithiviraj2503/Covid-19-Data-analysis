```python
import pandas as pd
import numpy as np

dataframe=pd.read_csv("conmod.csv").head(1085)
dataframe["age"].fillna( method ='ffill', inplace = True)
dataframe["gender"].fillna( method ='ffill', inplace = True)
dataframe["symptom"].fillna( method ='ffill', inplace = True)
dataframe["symptom"].fillna( method ='bfill', inplace = True)

dataframe.to_csv('conmod1.csv')
print("completed")
```
